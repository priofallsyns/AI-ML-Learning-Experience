```python
import pandas as pd
```

```python
d = {
    "Sales": [100000, 222000, 150000, 180000, 120000, 200000, 250000, 175000, 140000, 230000,
              260000, 190000, 160000, 210000, 240000, 170000, 130000, 220000, 200000, 180000],
    "City": ["Tampa", "Orlando", "Miami", "Jacksonville", "Tampa", "Orlando", "Miami", "Jacksonville",
             "Tampa", "Orlando", "Miami", "Jacksonville", "Tampa", "Orlando", "Miami", "Jacksonville",
             "Tampa", "Orlando", "Miami", "Jacksonville"],
    "Size": ["Small", "Medium", "Large", "Medium", "Small", "Large", "Medium", "Small", "Medium", "Small",
             "Large", "Medium", "Large", "Medium", "Small", "Large", "Medium", "Large", "Medium", "Small"]
}
```

```python
df=pd.DataFrame(data=d)
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')
```

```python
ohetransformed=ohe.fit_transform(df[['City']])
```

```python
ohetransformed.head()
```

```python
df=pd.concat([df, ohetransformed], axis=1).drop(columns=['City'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
sizes=df['Size'].unique()
print(sizes)
```

```python
oe=OrdinalEncoder(categories=[sizes], handle_unknown='use_encoded_value', unknown_value=-1)
```

```python
df['Size']=oe.fit_transform(df[['Size']])
```

```python
df.head()
```

```python
#we will use the first data 'd' we used in the start
```

```python
df=pd.DataFrame(data=d)
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
ohe=OneHotEncoder(sparse_output=False) #to make sure we can get the output as pandas DF
```

```python
ode=OrdinalEncoder()
```

```python
from sklearn.compose import make_column_transformer
```

```python
ct=make_column_transformer(
    (ohe, ['City']),
    (ode, ['Size']),
    remainder='passthrough'
)
```

```python
ct.set_output(transform='pandas')
```

```python
df_workedon=ct.fit_transform(df)
```

```python
df_workedon
```

```python
#now for the drop :
```

```python
ct2=make_column_transformer(
    (ohe, ['City']),
    (ode, ['Size']),
    remainder='drop'
)
```

```python
ct2.set_output(transform='pandas')
```

```python
df_dropped=ct2.fit_transform(df)
```

```python
df_dropped
```
