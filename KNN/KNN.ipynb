```python
import pandas as pd
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.metrics import confusion_matrix , classification_report
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('500hits.csv', encoding='latin-1')
```

```python
df
```

```python
###We Will Start by cleaning the data
```

```python
df=df.drop(columns=['PLAYER', 'CS'])
```

```python
X=df.iloc[:, 0:13]
```

```python
y=df.iloc[:, 13]
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
##now the data did drop the unnecessary columns , seperated the needed from the uneeded data, now we get to scaling the data
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.fit_transform(X_test)
```

```python
#Now the data is scaled , we can jump into the Model
```

```python
knn=KNeighborsClassifier(n_neighbors=8)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
print(y_test)
```

```python
print(y_pred)
```

```python
knn.score(X_test, y_test)
```

```python
cm=confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```
