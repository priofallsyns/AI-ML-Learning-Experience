```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_csv('Pearl_Jam_Tour2.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
df
```

```python
df=df.drop(columns=['city'])
```

```python
df2=pd.get_dummies(df[['continent', 'day_of_week']], dtype=int)
```

```python
df
```

```python
df
```

```python
df
```

```python
df
```

```python
df=pd.concat([df, df2], axis=1).drop(columns=['continent', 'day_of_week'], axis=1)
```

```python
df
```

```python
df['sold_out']=df['sold_out'].map({"yes":1, "no":0})
```

```python
df
```

```python
X=df.drop(columns=['sold_out'] , axis=1)
```

```python
y=df.iloc[:, 3:4]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
GNB=GaussianNB()
```

```python
GNB.fit(X_train, y_train)
```

```python
GNB.score(X_test, y_test)
```

```python
GNB.score(X_train, y_train)
```

```python
#####Parameters with GridSearch
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]
}
```

```python
grid_search = GridSearchCV(estimator=GNB, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```
