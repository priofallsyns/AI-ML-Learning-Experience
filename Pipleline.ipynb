```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = {
    "Social_media_followers": [1000000.0, None, 2000000.0, 1310000.0, 1700000.0, None, 4100000.0, 1600000.0, 2200000.0, 1000000.0],
    "Sold_out": [1, 0, 0, 1, 0, 0, 0, 1, 0, 1]
}
```

```python
df1=pd.DataFrame(data)
```

```python
df1
```

```python
X1=df1.iloc[:, 0:1]
```

```python
y1=df1.iloc[:, 1]
```

```python
y1
```

```python
X1
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.impute import SimpleImputer
```

```python
X1_train , X1_test, y1_train , y1_test=train_test_split(X1, y1, test_size=0.3, random_state=19 )
```

```python
si=SimpleImputer(strategy='mean')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR=LogisticRegression()
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe1=make_pipeline(si, LR)
```

```python
pipe1.fit(X1_train, y1_train)
```

```python
pipe1.score(X1_test, y1_test)
```

```python
###more advanced pipleline
```

```python
d2={
    "Genre": ['Rock', 'Metal', 'BlueGrass', 'Rock', np.nan , 'Rock', 'Rock' ,np.nan , 'BlueGrass', 'Rock'],
    "Social_media_followers": [1000000, np.nan ,2000000, 1310000, 1700000, np.nan , 4100000, 1600000, 2200000, 1000000],
    "Sold_out": [1,0,0,1,0,0,0,1,0,1]
}
```

```python
df=pd.DataFrame(d2)
```

```python
df
```

```python
X=df.iloc[:, 0:2]
```

```python
y=df.iloc[:, 2]
```

```python
X_train , X_test, y_train , y_test=train_test_split(X, y, test_size=0.3 , random_state=75)
```

```python
num_cols=['Social_media_followers']
```

```python
cat_cols=['Genre']
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe=Pipeline(steps=[
    ('impute', SimpleImputer(strategy='mean')),
    ('scale', StandardScaler() )
])
```

```python
cat_pipe=Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
col_tranformer=ColumnTransformer(transformers=[
    ('num_pipe', num_pipe, num_cols),
    ('cat_pipe', cat_pipe, cat_cols)
],
  remainder='drop',
  n_jobs=-1
                                 )
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
```

```python
pipefinal=make_pipeline(col_tranformer , dtc)
```

```python
pipefinal.fit(X_train, y_train)
```

```python
pipefinal.score(X_test, y_test)
```

```python
###Saving pipelines
```

```python
import joblib
```

```python
joblib.dump(pipefinal, 'pipe.joblib')
```
