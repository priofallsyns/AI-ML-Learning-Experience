```python
import pandas as pd
```

```python
import numpy as np
```

```python
np.random.seed(42)
```

```python
fastball_speed=np.random.randint(90, 106, size=500)
```

```python
tommy_john=np.where(fastball_speed>96, np.random.choice([0,1], size=500, p=[0.3, 0.7]), 0)
```

```python
data={
    "fastball_speed": fastball_speed,
    "tommy_john": tommy_john
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.head(500)
```

```python
X=df[['fastball_speed']]
```

```python
y=df[['tommy_john']]
```

```python
y
```

```python
####Normal method
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=25)
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_test , y_test)
```

```python
##basic cross validation
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cvscore=cross_val_score(lr, X, y, cv=10 )
```

```python
cvscore
```

```python
np.max(cvscore)
```

```python
from sklearn.model_selection import KFold
```

```python
kf=KFold(n_splits=15, shuffle=True, random_state=42)
```

```python
cvscore=cross_val_score(lr, X, y, cv=kf)
```

```python
np.max(cvscore)
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
SFK=StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
```

```python
csv=cross_val_score(lr , X, y, cv=SFK)
```
